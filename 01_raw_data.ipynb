```python
import pandas as pd
import pyarrow
```

# Data load

```python
df_raw_2025 =  pd.read_csv("data/raw/IMP_2025.csv", delimiter=";")
df_raw_2024 =  pd.read_csv("data/raw/IMP_2024.csv", delimiter=";")
df_raw_2023 =  pd.read_csv("data/raw/IMP_2023.csv", delimiter=";")
df_raw_2022 =  pd.read_csv("data/raw/IMP_2022.csv", delimiter=";")
df_raw_2021 =  pd.read_csv("data/raw/IMP_2021.csv", delimiter=";")
df_raw_2020 =  pd.read_csv("data/raw/IMP_2020.csv", delimiter=";")

# Empilhamento do hist√≥rico
df_list = [df_raw_2025, df_raw_2024, df_raw_2023, df_raw_2022, df_raw_2021, df_raw_2020]
df_raw = pd.concat(df_list, ignore_index=True)
```
